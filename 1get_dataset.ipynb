- Este df tiene algún problema respecto a las columnas de latitude y longitude
- Hacemos una normalización previa para que estén igual que los otros tres datasets

```python
import pandas as pd
import numpy as np

def normalize_coord(coord, is_latitude=True):

    try:
        num = float(coord)
        if num == 0:
            return num

        abs_num = abs(num)
        sign = -1 if num < 0 else 1

        # Determinar la posición del punto decimal basado en el orden de magnitud
        if is_latitude:
            # Latitud: esperamos valores ~40 (2 dígitos enteros)
            if abs_num > 999:
                return sign * (abs_num / 1000)
            elif abs_num > 99:
                return sign * (abs_num / 100)
            elif abs_num > 9:
                return sign * (abs_num / 10)
        else:
            # Longitud: esperamos valores ~-3.7 (1 dígito entero)
            if abs_num > 999:
                return sign * (abs_num / 1000)
            elif abs_num > 99:
                return sign * (abs_num / 100)
            elif abs_num > 9:
                return sign * (abs_num / 10)

        return num
    except:
        return np.nan
```

```python

# Aplicar a tu DataFrame
df['latitude'] = df['latitude'].apply(lambda x: normalize_coord(x, is_latitude=True))
df['longitude'] = df['longitude'].apply(lambda x: normalize_coord(x, is_latitude=False))
df['latitude'] = df['latitude'].apply(lambda x: normalize_coord(x, is_latitude=True))
df['longitude'] = df['longitude'].apply(lambda x: normalize_coord(x, is_latitude=False))
```

- creo columnas nuevas:

```python
import numpy as np

# Función para mapear exterior/interior
def map_exterior(desc):
    if pd.isna(desc):
        return np.nan
    desc = str(desc).lower()
    if 'exterior' in desc:
        return 1
    elif 'interior' in desc:
        return 0
    else:
        return np.nan

# Función para mapear ascensor
def map_ascensor(desc):
    if pd.isna(desc):
        return np.nan
    desc = str(desc).lower()
    if 'con ascensor' in desc:
        return 1
    elif 'sin ascensor' in desc:
        return 0
    else:
        return np.nan


df_final['exterior'] = df_final['floor_description'].apply(map_exterior)
df_final['ascensor'] = df_final['floor_description'].apply(map_ascensor)

print("Distribución de valores para 'exterior':")
print(df_final['exterior'].value_counts(dropna=False))
print("\nDistribución de valores para 'ascensor':")
print(df_final['ascensor'].value_counts(dropna=False))
```

```python
import re
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier

def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text).lower() #minusculas
    text = re.sub(r'[^\w\s]', '', text)  # Elimina puntuación
    return text

df_final['cleaned_desc'] = df_final['property_description'].apply(clean_text)
```

```python
vectorizer = TfidfVectorizer(
    max_features=500,
    stop_words=['el', 'la', 'de', 'en', 'y', 'con']
)
X = vectorizer.fit_transform(df_final['cleaned_desc'])
```

```python
from sklearn.model_selection import train_test_split

train_data = df_final.dropna(subset=['floor'])
X_train = vectorizer.transform(train_data['cleaned_desc'])
y_train = train_data['floor'].astype(str)  # Convertir a string para clasificación

#clasificación (ajusta hiperparámetros si es necesario)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

#valores faltantes - busco filas con na y aplico solo a ella
null_indices = np.where(df_final['floor'].isna())[0]

if len(null_indices) > 0:
    X_null = X[null_indices]  #solo filas nulas
    df_final.loc[df_final.index[null_indices], 'floor'] = model.predict(X_null)

print(f"Valores imputados: {df_final['floor'].isna().sum()} nulos restantes")
print("Distribución de valores:\n", df_final['floor'].value_counts(dropna=False))
```

```python

import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

def impute_binary_column(col_name, positive_words, negative_words):
    y = df_final[col_name].copy()

    mask_pos = df_final['property_description'].str.contains(
        '|'.join([re.escape(word) for word in positive_words]), na=False, case=False
    )
    mask_neg = df_final['property_description'].str.contains(
        '|'.join([re.escape(word) for word in negative_words]), na=False, case=False
    )

    y = np.where(
        df_final[col_name].notna(),
        df_final[col_name],
        np.where(mask_pos, 1,
                 np.where(mask_neg, 0, np.nan))
    )

    train_mask = df_final[col_name].notna()
    train_indices = np.where(train_mask)[0]

    null_mask = df_final[col_name].isna() & np.isnan(y)
    null_indices = np.where(null_mask)[0]

    if len(train_indices) > 0:
        print("aplicando classifier")
        model = RandomForestClassifier(n_estimators=100, random_state=42)
        model.fit(X[train_indices], y[train_indices].astype(int))

        if len(null_indices) > 0:
            df_final.loc[df_final.index[null_indices], col_name] = model.predict(X[null_indices])

    print(f"\n=== Reporte para {col_name} ===")
    if len(train_indices) > 0:
        print(classification_report(y[train_indices], model.predict(X[train_indices])))
    else:
        print("No hay datos etiquetados para entrenar el modelo.")


#palabras clave para cada variable para buscarlas lo primero
exterior_words = {
    'positive': ['exterior', 'fachada', 'balcón', 'terraza'],
    'negative': ['interior', 'patio interior']
}

ascensor_words = {
    'positive': ['ascensor', 'elevador'],
    'negative': ['sin ascensor', 'no ascensor']
}

# 5. Ejecutar imputación
impute_binary_column('exterior', exterior_words['positive'], exterior_words['negative'])
impute_binary_column('ascensor', ascensor_words['positive'], ascensor_words['negative'])

# 6. Resultados finales
print("\nDistribución de valores:")
print("exterior:\n", df_final['exterior'].value_counts(dropna=False))
print("\nascensor:\n", df_final['ascensor'].value_counts(dropna=False))
```
